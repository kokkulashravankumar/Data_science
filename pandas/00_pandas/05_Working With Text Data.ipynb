```python
'''In order to lowercase a data, we use str.lower() this function converts all uppercase characters to lowercase. 
If no uppercase characters exist, it returns the original string. 
In order to uppercase a data, we use str.upper() this function converts all lowercase characters to uppercase.
If no lowercase characters exist, it returns the original string.'''

# Import pandas package 
import pandas as pd 
   
# Define a dictionary containing employee data 
data = {'Name':['JAi', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']} 
   
# Convert the dictionary into DataFrame  
df = pd.DataFrame(data) 
   
# converting and overwriting values in column 
df["Name"]= df["Name"].str.lower()
 
print(df)
```
