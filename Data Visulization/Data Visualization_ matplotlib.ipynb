# Colors,Marker& Fonts

```python
import matplotlib.pyplot as plt
#the number views  a youtube channel got in last 7 days

views=[534,689,258,401,724,689,350]
days=range(1,8)#[1,2,3,4,5,6,7,8] we aslo speficed by this
#plt.plot(x,y)
plt.plot(days,views,label='youtube views',color='r',marker='o',markerfacecolor='g')
plt.xlabel('no_days')
plt.ylabel('views')
plt.legend()
plt.title('youtube views on daily basics')
plt.show()
```

# Line width &  style

```python
import matplotlib.pyplot as plt
#the number views  a youtube channel got in last 7 days

views=[534,689,258,401,724,689,350]
days=range(1,8)#[1,2,3,4,5,6,7,8] we aslo speficed by this
#plt.plot(x,y)
plt.plot(days,views,label='youtube views',color='r',marker='o',markerfacecolor='g',linestyle='--',linewidth=3)
plt.xlabel('no_days')
plt.ylabel('views')
plt.legend()
plt.title('youtube views on daily basics')
plt.show()
```
